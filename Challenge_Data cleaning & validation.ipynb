```python
import pandas as pd
from collections import Counter
import numpy as np
from IPython.display import HTML, display
# import tabulate
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding='iso-8859-1')
df.head()
```

```python
cost = df.iloc[:, 4].str.replace('£', '')
cost = cost.str.replace('$', '')
cost = pd.to_numeric(cost) # convert string to float
print(cost.mean())
print(cost.median())
print(cost.max())
print(cost.std())
df.iloc[:, 4] = cost
capitalizer = lambda x:str(x).upper()
df['Journal title']=df['Journal title'].apply(capitalizer)
df = df[np.abs(df.iloc[:, 4]- cost.mean())<=(3*cost.std())] # exclude extreme values
df.head()
```

```python
journal_counts = Counter(df['Journal title'])
rank_result = pd.DataFrame.from_dict(journal_counts, orient='index')
num_items = 5  # show top #num_items rows 
rank_result.columns = ['count'] # set name to column for convenience
rank_result.sort_values('count',inplace=True, ascending=False)
journals_info = rank_result[0:num_items].copy()
print(journals_info)
for journal in journals_info.index:
    print(journal)
    top_five = df[df['Journal title']==journal]
    display(top_five.head())
    print(top_five.iloc[:, 4].mean())
    print(top_five.iloc[:, 4].median())
    print(top_five.iloc[:, 4].max())
    print(top_five.iloc[:, 4].std())
```
